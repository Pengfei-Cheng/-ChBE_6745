### 2. Proposed method ###

```python
import numpy as np
import matplotlib.pyplot as plt

def un(x, n):
    if x >= 1/n:
        return 1 / n**2
    else:
        return -1 / n**2

def fk(xspan, k):
    yspan = np.zeros(xspan.shape)
    for j in range(1, k+1):
        for i, x in enumerate(xspan):
            yspan[i] += un(x, j)
            
    return yspan
    
xspan = np.linspace(0, 2)

    
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes[0][0].plot(xspan, fk(xspan, 1))
axes[0][1].plot(xspan, fk(xspan, 2))
axes[1][0].plot(xspan, fk(xspan, 3))
axes[1][1].plot(xspan, fk(xspan, 4))
```
