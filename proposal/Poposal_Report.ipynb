```python
import numpy as np
import matplotlib.pyplot as plt

def un(x, n):
    if x >= 1/n:
        return 1 / n**2
    else:
        return -1 / n**2

def fk(xspan, k):
    yspan = np.zeros(xspan.shape)
    for j in range(1, k+1):
        for i, x in enumerate(xspan):
            yspan[i] += un(x, j)
            
    return yspan
    
xspan = np.linspace(0, 2, 1000)

    
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes[0][0].scatter(xspan, fk(xspan, 1))
axes[0][0].set_title("$f_1$(x)")
axes[0][1].scatter(xspan, fk(xspan, 2))
axes[0][1].set_title("$f_2$(x)")
axes[1][0].scatter(xspan, fk(xspan, 3))
axes[1][0].set_title("$f_3$(x)")
axes[1][1].scatter(xspan, fk(xspan, 4))
axes[1][1].set_title("$f_4$(x)")
```
